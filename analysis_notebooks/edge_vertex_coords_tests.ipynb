Plots Map of table with labels associated to each coordinate / location on steelplate

```python
#* for data manipulation 
import pandas as pd 
from disk_analysis_tools import tiling_disk_plots as tdp

#* for plotting: 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
sns.set_theme(style="ticks")
sns.set_context("talk")

from pathlib import Path
measurement_folder  = Path.cwd().parent / 'excel_data'
```

```python
coord_data_edges = pd.read_excel(measurement_folder/'coords_xy_edges_verticies.xlsx', sheet_name='coordinates')
```

```python
coord_data_edges
```

```python
tdp.coordinates_plot(coord_data_edges)
ax = plt.gca()
for i, row in coord_data_edges.iterrows():
    if row['arm_type'] == 'vertex':
        ax.annotate(row['arm_label'], (row['x'], row['y']), color='tab:orange',xycoords='data')
    else:
        ax.annotate(row['arm_label'], (row['x'], row['y'])) 
```
